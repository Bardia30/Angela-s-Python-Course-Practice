```python
#BMI 2.0 
height = float(input("enter your height in m: "))
weight = float(input("enter your weight in kg: "))
bmi = round(weight/(height)**2)

if bmi <= 18.5:
    print("your BMI is ",bmi,"you are underweight.")
elif bmi >18.5 and bmi <= 25:
    print("your BMI is ",bmi,"you are normal weight")
elif bmi > 25 and bmi <= 30:
    print("your BMI is ",bmi,"you are overweight")
elif bmi >30 and bmi < 35:
    print("your BMI is ",bmi,"you are obese")
else:
    print("your BMI is",bmi,"you are clinically obese")
    


```

```python
#pizza order
print("welcome to Python Pizza Deliveries!")
size=input("what size pizza do you want?S,M,L?: ")
add_pepperoni = input("Do you want pepperoni?: ")
extra_cheese = input("Do you want extra cheese?: ")

price=0

if size == "S":
    price = 15
    if add_pepperoni == "Y":
        price = price+ 2
elif size == "M":
    price = 20
    if add_pepperoni == "Y" :
        price +=3
elif size == "L":
    price = 25
    if add_pepperoni == "Y" :
        price +=3
else:
    print("invalid input")


if extra_cheese == "Y":
    price+=1

print("your final bill is: $",price)

```

```python
#Love Calculator

print("Welcome to the Love Calculator!")
name1 = input("What is your name? \n").lower()
name2 = input("What is their name? \n").lower()

name_combine = name1 + name2

t_count = name_combine.count("t")
r_count = name_combine.count("r")
u_count = name_combine.count("u")
e_count_1 = name_combine.count("e")
first_digit = str(t_count + r_count + u_count + e_count_1)


l_count = name_combine.count("l")
o_count = name_combine.count("o")
v_count = name_combine.count("v")
e_count_2 = name_combine.count("e")
second_digit = str(l_count + o_count + v_count + e_count_2)

final_value = int(first_digit + second_digit)

if final_value <= 10 or final_value >= 90:
    print("Your score is",final_value, ", you go together like coke and mentos.")
elif final_value >= 40 and final_value <= 50:
    print("Your score is",final_value, ", you are alright together.")
else:
    print("your score is ",final_value)


```
