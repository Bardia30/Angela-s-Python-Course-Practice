```python
#calculator
import art_calc


#add 
def add(n1, n2):
    return n1 + n2

#subtract
def subtract(n1,n2):
    return n1 - n2

#multiply
def multiply(n1,n2):
    return n1*n2


#divide
def divide(n1,n2):
    return n1/n2

operations ={"+":add,
             "-" : subtract,
             "*" : multiply, 
             "/": divide
}


def calculator():
    print(art_calc.logo)
    num1  = float(input("what is the first number?: "))




    for key in operations: 
        print(key)

    run=False
    while not run:

        operation_symbol = input("which of these opearations you want?")

        num2  = float(input("what is the second number?: "))


        calc_func = operations[operation_symbol]

        answer = calc_func(num1,num2)

        print(num1,operation_symbol,num2,"=",answer)

        running = input(f"Type 'y' to continue calculating with {answer} or type 'n' to run a new calculation, or 'exit' to finish the calculator: ")
        if running == "y":
            run = False
            num1=answer
        elif running == "n":
            run = True
            calculator()
        elif running == "exit":
            break

calculator()
```
