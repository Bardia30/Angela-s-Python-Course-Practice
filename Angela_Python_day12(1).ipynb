```python
#modifying a global variable
enemies = 1

def increase_enemies():
    global enemies #searches for global variable anywhere outside the function
    enemies += 1
    print("Your enemies are: ",enemies)
    
increase_enemies()
print("Your enemies are: ",enemies)
```

```python
#Angela's solution
from random import randint

easy_level_turns = 10
hard_level_turns = 5


def check_answer(guess,answer,turns):
    if guess > answer:
        print("too high.")
        return turns-1
    elif guess < answer:
        print("too low.")
        return turns-1
    else:
        print(f"you got it! The answer was {answer}")

        
        
        
def set_difficulty():
    level = input("choose a difficulty level. easy or hard: ")
    if level == "easy":
        return easy_level_turns
    else:
        return hard_level_turns

def game():
    print("welcome to the guessing game")
    print("Im thinking of a number between 1 and 100")
    answer = randint(1,100)
    print(f"pst the answer is {answer}")
    turns = set_difficulty()
    


    guess=0
    while guess != answer:
        print(f"you have {turns} attempts remaining")
        guess = int(input("make a guess:  "))
        turns = check_answer(guess, answer,turns)
        if turns == 0:
            print("you've run out of guesses. You lose")
            return #this makes the function to stop 
        elif guess !=answer:
            print("Guess again.")


game()

```
